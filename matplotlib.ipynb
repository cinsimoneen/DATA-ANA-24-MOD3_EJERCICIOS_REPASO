```python
#Histograma básico
import numpy as np
import matplotlib.pyplot as plt

numeros=np.random.randn(1000)
plt.hist(numeros, bins=40, alpha=0.5, color='orange',edgecolor='gray')  #bins, cuantas columnas queremos que se vean, alpha=transparencia del color
#color es color de fondo, edgecolor es color del borde, histtype el tipo de histagrama que queremos por ejemplo step o stepfilled
```

```python
#Histograma con varios histogramas superpuestos
import numpy as np
import matplotlib.pyplot as plt

x1=np.random.randn(1000)
x2=np.random.randn(500)
x3=np.random.randn(100)

plt.hist(x1, bins=40, histtype='stepfilled',alpha=0.3)
plt.hist(x2, bins=40, histtype='stepfilled',alpha=0.3)
plt.hist(x3, bins=40, histtype='stepfilled',alpha=0.3)
```

```python
# graficas de pie
import matplotlib.pyplot as plt
import pandas as pd

df=pd.DataFrame({"etiquetas":["Manzanas","Naranjas","Bananas","Uvas"],
                 "cantidad":[30,60,80,20]})
df
plt.pie(df['cantidad'],labels=df['etiquetas'],autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40E0D0'],
        explode=(0,0,0,0.5))
#autopct nos calcula el porcentaje automaticamente, colors cambia los colores de default, explode separa la rebanada en este caso uva
plt.title("Mis frutas")
```

```python
#Graficos Multiples
import matplotlib.pyplot as plt
%matplotlib inline

fig,axs=plt.subplots(ncols=2,nrows=1)
ax1=axs[0]
ax2=axs[1]

ax1.plot([1,2,3,4],[1,4,9,16])
ax2.scatter([1,2,3,4],[1,4,9,16])

ax1.set_title('Gráfico de línea')
ax2.set_title('Gráfico scatter')
;
```
