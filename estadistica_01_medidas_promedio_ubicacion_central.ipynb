```python
# Media Aritmética   #suma de todos los datos/numero de datos
#La media aritmética es la suma de todos los datos dividida por el número de datos. 
# Comúnmente, se conoce como "promedio", aunque hay muchos promedios matemáticos distintos. 

#La media, o promedio aritmético, se calcula sumando todos los valores del conjunto de datos y dividiendo el resultado entre 
# el número total de valores. La media ofrece una idea del valor central de un conjunto de datos. 
# Sin embargo, es sensible a los valores atípicos (outliers), que pueden sesgar la media significativamente

import statistics
import numpy as np
from scipy import stats
import pandas as pd

data_df = pd.DataFrame({
    "col1": [1,2,3,2,2,100],
    "col2": [4.,5,6,7,8,9],
    "col3": [3,3,2,2,6,6]    
})
#data_df
# poniendo una columna del df en una serie 
data_serie=data_df.col2
data_serie

# poniendo la serie en un arreglo numpy 
data_vector=data_serie.to_numpy()
data_vector

statistics.mean(data_serie) # si puede calcular la media cuando es una serie
statistics.mean(data_vector) #si puede calcular la media cuando es un arreglo numpy
#statistics.mean(data_df) # usando statistics no puede calcular la media de un dataframe directo

#calcular la media de todo un dataframe usando lambda
data_df.apply(lambda x: statistics.mean(x))

##calculo con scipy
stats.tmean(data_serie)  #tmean nos deja poner limites como en el ejemplo de abajo
stats.tmean(data_vector)
stats.tmean(data_df)  # aqui calcula la media de las tres columnas
#stats.tmean(data_df.col1, limits=(0,5)) #para que no tome el 100 que es un valor no concordante

#calculo con numpy
np.mean(data_serie)  #calcula sin problema
np.mean(data_vector) #calcula sin problema
np.mean(data_df)  #calcula sin problema las tres columnas

#calculo de pandas
data_serie.mean()
#esto es de numpy
data_vector.mean()

np.mean(data_df) 
```

```python
# MODA
#La moda es el valor que más veces se repite en el conjunto de datos, es decir, el valor más frecuente.

#La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. 
# Un conjunto de datos puede tener una moda (unimodal), más de una moda (bimodal o multimodal) 
# o no tener ninguna moda si todos los valores son únicos. 
# Es útil para identificar el valor más común en un conjunto de datos, 
# proporcionando una idea de la concentración de datos en torno a ciertos valores específicos.

import statistics
import numpy as np
from scipy import stats
import pandas as pd

data_df = pd.DataFrame({
    "col1": [1,2,3,2,2,100],
    "col2": [4.,5,6,7,8,9],
    "col3": [3,3,2,2,6,6]    
})

# poniendo una columna del df en una serie 
data_serie=data_df.col2
data_serie

# poniendo la serie en un arreglo numpy 
data_vector = data_serie.to_numpy()
data_vector

#usando statistics
statistics.mode(data_df.col1)
statistics.mode(data_df.col2) # cuando hay mas de una moda devuelve la primera
statistics.mode(data_df.col3) # cuando hay mas de una moda devuelve la primera

statistics.multimode(data_df.col2)  #devuelve todas las modas
statistics.multimode(data_df.col3)

#con scipy devuelve solo la que tenga el valor mas pequeño
stats.mode(data_df.col2)
stats.mode(data_df.col3) # devuelve el valor de la moda y cuantas veces aparece

# usando numpy  no l o tiene

# con pandas
data_serie.mode()
data_df.mode() #aqui devuelve modas que encontró en cada una de las columnas
```
