```python
# metodo iloc  por indice
#Seleccion de filas y columnas
import pandas as pd

data_diabetes=pd.read_csv(filepath_or_buffer="tmp/diabetes_data_upload.csv")
df=pd.DataFrame(data_diabetes)
#df

#seleccionar una fila con iloc
#df.iloc[0:20] #trae las 20 primeras filas
df.iloc[500] # trae la fila con indice 500

#seleccionar las columnas con iloc
df.iloc[:,2] # trae la columna con indice 2 pero trae  filas por el primer parametro asi lo dice

#seleccionar filas y columnas
df.iloc[0:3,2] #aqui pone de la primera fila a la 3 sin incluir solo los valores de la columna 2
df.iloc[:,0:3] #todas las filas de las columnas de la cero a la 3
df.iloc[[0,1,5],:] #las filas con indice 0, 1 y 5 todas las columnas
df.iloc[[0,1,5],[0,2,3]] #las filas con indice 0, 1 y 5 las columnas 0,2 y 3
```

```python
# metodo loc  por etiquetas
#Seleccion de filas y columnas
import pandas as pd

data_diabetes=pd.read_csv(filepath_or_buffer="tmp/diabetes_data_upload.csv")
df=pd.DataFrame(data_diabetes)
df

#usando etiquetas en vez de nombres
df.loc[:,['<Age','Gender','Alopecia']] #todas las filas de esas tres columnas
df.loc[30,"Gender"] #trae el valor del indice 30 para la columna Gender
df.loc[0:10,['<Age','Gender','Alopecia']]# a diferencia de iloc, loc inluye la fila 10 
```

```python
#Seleccion de filas y columnas con condiciones
import pandas as pd

data_diabetes=pd.read_csv(filepath_or_buffer="tmp/diabetes_data_upload.csv")
df=pd.DataFrame(data_diabetes)
df

df.loc[:,"Gender"]=="Male"  #aqui traemos tru o false si cumple o no con la busqueda
df.loc[df.loc[:,"Gender"]=="Male"]  #aqui traemos ya filtrada la busqueda
df.loc[:,"Gender"].unique() #metodo unique es distint en sql
df.loc[(df.loc[:,"Gender"]=="Male")  & (df.loc[:,'<Age']>50)]  #aqui trae las dos condiciones hombres mayores de 50
# operadores | or 

#otra forma sin el loc y solo indexando con pandas seria
df[(df["Gender"]=="Male") & (df['<Age']>50)]  #aqui trae las dos condiciones hombres mayores de 50

# isin
lista=[50,55]
df.loc[df.loc[:,'<Age'].isin(lista)] # traernos los registros que cumplen con 50 y 55( lista)

```

```python
#insertar columnas
#Seleccion de filas y columnas con condiciones
import pandas as pd

data_diabetes=pd.read_csv(filepath_or_buffer="tmp/diabetes_data_upload.csv")
df=pd.DataFrame(data_diabetes)


df.loc[:,"class"]="col_class"  #asignando a class un valor constante igual a col_class
#sustituyendo con condicion
df.loc[df["Gender"]=="Male","weakness"] ="Si"  # cambiando valor con condicion, si es Male cambia otra columna weakness a si
#cambiando dos columnas
df.loc[df["Gender"]=="Male",["weakness","Alopecia"]] ="Si","Probable"  # cambiando valor con condicion, si es Male cambia otra columna weakness a si

df
```
